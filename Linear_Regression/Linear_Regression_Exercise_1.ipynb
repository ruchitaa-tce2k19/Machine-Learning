Predict canada's per capita income in year 2020. There is an exercise folder here on github at same level as this notebook, download that and you will find canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'D:\Machine Learning\py-master\py-master\ML\1_linear_reg\Exercise\canada_per_capita_income.csv')
df               
```

```python
df.columns
```

```python
df.columns=['year','per_capita_income']
```

```python
df.columns
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year,df.per_capita_income,color='red',marker='+')
```

```python
new_df=df.drop('per_capita_income',axis='columns')
new_df
```

```python
income=df.per_capita_income
income
```

Creating linear regression object

```python
reg=linear_model.LinearRegression()
reg.fit(new_df,income)
```

Predict the per capita income fo canadian citizens in year 2020

```python
print('Per capita income fo canadian citizens in year 2020')
reg.predict([[2020]])
```
