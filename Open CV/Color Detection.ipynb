We are going to detect colors in a given image. The picture that I have taken has 2 colors, red and black. Let us detect these 2 colors in the image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('D:\wall paper\pic.jpg')
```

```python
plt.figure(figsize=(20,8))
plt.imshow(img)
```

By default open CV loads the image in BGR color format. We have to convert it into RGB format.

```python
grid_RGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,8))
plt.imshow(grid_RGB)
```

HSV - (hue, saturation, value), also known as HSB (hue, saturation, brightness), is often used by artists because it is often more natural to think about a color in terms of hue and saturation than in terms of additive or subtractive color components. HSV is a transformation of an RGB colorspace, and its components and colorimetry are relative to the RGB colorspace from which it was derived.

```python
grid_HSV = cv2.cvtColor(grid_RGB,cv2.COLOR_RGB2HSV)
```

Inorder to detect a color, we have to give color range. We have to mention lower and higher HSV color range.

DETECTING RED COLOR

```python
lower = np.array([0,150,50])
upper = np.array([10,255,255])
```

```python
mask = cv2.inRange(grid_HSV,lower,upper)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

Bitwise operations are used in image manipulation and used for extracting essential parts in the image

```python
res = cv2.bitwise_and(grid_RGB,grid_RGB,mask=mask)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(res)
```

DETECTING BLACK COLOR

```python
lower1 = np.array([0,0,0])
upper1 = np.array([250,255,30])
```

```python
mask = cv2.inRange(grid_HSV,lower1,upper1)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

```python
res1 = cv2.bitwise_and(grid_RGB,grid_RGB,mask=mask)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(res1)
```
